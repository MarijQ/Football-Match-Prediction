```python
import pandas as pd

df = pd.read_csv('clubs.csv')
df
```

```python
#check the data type of the dataset
df.dtypes
```

```python
#checking duplicates

df.duplicated().sum()
```

```python
#finding null values in the dataset

df.isnull().sum()
```

```python
from unidecode import unidecode

df['name'] = df['name'].apply(unidecode)
```

```python
#dropping the unnecessary columns
'''
1 - club_code
2 - name
3 - squad_size
4 - foreign_numbers
5 - foreign_percentage
6 - national_team_players
7 - coach_name
8 - last_season
9 - filename
10 - url
11 - total_market_value
'''

df = df.drop(['club_code','name', 'squad_size', 'foreigners_number', 'foreigners_percentage','national_team_players', 'coach_name', 'last_season', 'filename', 'url', 'total_market_value'], axis=1)
```

```python
#converting unicode to correct data

df['stadium_name'] = df['stadium_name'].apply(unidecode)
```

```python
#cleaning 'net_transfer_record'

#renaming 'net_transfer_record' to 'net_transfer_record_in_euros'
df.rename(columns={'net_transfer_record': 'net_transfer_record_in_euros'}, inplace=True)

#removing '€' by replacing it with an empty string

df['net_transfer_record_in_euros'] = df['net_transfer_record_in_euros'].str.replace('€', '')

#cheching the data type

df['net_transfer_record_in_euros'].dtype

#converting the data type
'''

Need to need to convert the data type

'''

```

```python
#replacing null values for 'average_age' with the random impution
import random

random_age = random.choice(df['average_age'])
random_age

df['average_age'].fillna(random_age, inplace=True)
```

```python
df.to_csv('clubs_cleaned.csv', index=False)
```
